```python
%load_ext sql
conn = duckdb.connect()
%sql conn --alias duckdb
%config SqlMagic.displaylimit = 50
```

```python
math_xls = pd.read_excel("25Math03SRSSpring23.xls", skiprows=4, usecols="A:H")
duckdb.sql("CREATE TABLE math_xls AS SELECT * FROM math_xls")
math_xls.iloc[0].to_dict()
```

```python
%%sql
describe
select *
FROM math_xls
```

```python
%%sql

select "District Name", "School Name", "Mean  Scale Score "
from math_xls
where "District Name" = 'HILLSBOROUGH'
order by "Mean  Scale Score " desc
```
